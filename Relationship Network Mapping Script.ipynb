```python
from pyvis.network import Network
import pandas as pd

data = pd.read_csv("Team Relationship Mapping Survey (1-19).csv")
got_net = Network(height='99%', width='99%', directed=True)
got_net.set_edge_smooth('dynamic')

# Create the dataframe

data2 = data.melt(id_vars = "name", var_name="relation")

for index, row in data.iterrows():
    name = row[0]
    got_net.add_node(name, title=name, fontsize=90)

for index, row in data2.iterrows():
    name = row[0]
    relation = row[1]
    weight = row[2]

    #if weight == 1:
        #got_net.add_edge(name, relation, width=weight*5, title="I donâ€™t work with this person (0-2hrs a week)", color='#eb7005')
    #if weight == 2:
        #got_net.add_edge(name, relation, width=weight*5, title="I rarely work with this person (2-5hrs a week)", color='#c2885b')
    #if weight == 3:
        #got_net.add_edge(name, relation, width=weight*5, title="I sometimes work with this person (6-7hrs a week)", color='#708090')
    #if weight == 4:
        #got_net.add_edge(name, relation, width=weight*5, title="I regularly work with this person (8-10hrs a week)", color='#2a7ab0')
    if weight == 5:
        got_net.add_edge(name, relation, width=weight*5, title="I often work with this person (10+hrs a week)", color='#004980')
        
got_net.set_options("""
var options = {
  "nodes": {
    "font": {
      "size": 82,
      "strokeWidth": 22
    },
    "size": 48
  },
  "edges": {
    "color": {
      "inherit": true
    },
    "smooth": {
      "forceDirection": "none"
    }
  },
  "physics": {
    "barnesHut": {
      "gravitationalConstant": -80000,
      "springLength": 250,
      "springConstant": 0.001
    },
    "minVelocity": 0.75
  }
}

""")




#got_net.barnes_hut(gravity=-80000, central_gravity=0.3, spring_length=250, spring_strength=0.001, damping=0.09, overlap=0)
#got_net.show_buttons()
got_net.show('TeamWorking5.html')
```
